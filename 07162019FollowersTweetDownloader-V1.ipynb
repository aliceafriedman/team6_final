```python
def get_users_tweets_json(username):
    # create empty dictionary to accept data
    dict_fol = {'created_at': [], 'verified': [], 'location': [], 'followers_count': [], 'tweet_text': [], 'retweet_count': [], 'favorite_count': []}  
    next_cursor = -1
    tweets = []
    result = ""
    while(next_cursor):
        try:
            result = twitter.get_user_timeline(screen_name=username, count = 200, cursor=next_cursor)
            for tweet in result:
                tweets.append(tweet)
            try: 
                time.sleep(5)
                next_cursor = result["next_cursor"]
                print("Getting more of %s's followers' tweets" % username)
                print (time.asctime())
            except:
                break
        except:
            f=open("ignore.txt","a+")
            f.write(username + "\n")
            f.close()
            break
    filename=username + "#" + str(time.time()) + "_tweets.json"
    
    with open("tweets/" + filename, "w") as file:
        json.dump(tweets, file)
        
    print("Saving %s's tweets to JSON file" % username)
    print (time.asctime())
    return result
```

Now that we have information about the major hashtags so lets construct a list of these words to find all the tweets related to politics

```python
seed_words = ['#DemDebate', '#YangGang', '#Bernie2020','#DemDebate2','#Iran','#MedicareForAll','#Trump','#LetYangSpeak','#DemocraticDebate','#Yang2020','#MAGA','#Yemen','#CancelStudentDebt','#nhpolitic','#Israel','#TrumpParade','#Venezuela','#BernieSander','#POTUS','#AndrewYang','#Syria','#YangOnTheView','#FeelTheBern','#BackfireTrump','#Saudi','#Iraq']
#Lets do some pre-processing on the tweets dataset
cleaned_tweets = []
for tweet in tweets:
    tweet = re.sub(r"http\S+", "", tweet) # remove links from tweet
    cleaned_tweets.append(tweet)
#Lets get all the tweets which have these hashtags
seed_tweets = []
for tweet in cleaned_tweets:
    if re.compile('|'.join(seed_words),re.IGNORECASE).search(tweet):
        seed_tweets.append(tweet)
```

```python
def jaccard(tweet1,tweet2):
    intersec = set(tweet1).intersection(set(tweet2))
    uni = set(tweet1).union(set(tweet1))
    return float(float(len(intersec))/float(len(uni)))
```

```python
similarity = 0.7 # anything more than 70% similar is good
selected_tweets = seed_tweets
cleaned_tweets = list(set(cleaned_tweets) - set(selected_tweets))# will remove duplicate tweets too
for stweet in seed_tweets:
    doc1 = [w for w in stweet.lower().split(" ") if not w in STOPWORDS] # remove stop words from current selected tweet
    for ctweet in cleaned_tweets:
        doc2 = [w for w in ctweet.lower().split(" ") if not w in STOPWORDS]# remove stop words from current clean tweet
        if jaccard(doc1,doc2) > similarity:
            selected_tweets.append(ctweet)
selected_tweets = list(set(selected_tweets))
```
